```python
# read

with open('01.txt', 'r') as f:
  data = f.read().split('\n')
```

```python
# part 1
# correct answer: 54390

# this does not work if one of the strings has no digits in it. thankfully my puzzle didn't 
# meet that condition and i don't care enough to fix it.

def calc_total(data):
  return sum([int(''.join(list(filter(str.isdigit, x))[e] for e in (0, -1))) for x in data])

print(calc_total(data))
```

```python
# part 2
# correct answer: 54277

# basic idea: read the string until the first digit is found.
# then, reverse the string and read again until the first digit is found. 

nums = (
  ('one', '1'), ('two', '2'), ('three', '3'), ('four', '4'), ('five', '5'),
  ('six', '6'), ('seven', '7'), ('eight', '8'), ('nine', '9')
)
cals = [{ 'start': 0, 'end': 0 } for _ in range(len(data))]

def find_first_digit(line, reverse=False, pos=0):
  window = ''

  if pos+4 <= len(line): window = line[pos:pos+5]
  elif (len(line) < 5): window = line
  else: return

  [window := window.replace(a[::-1] if reverse else a, b) for a, b in nums]
  digits = [w for w in window if w.isdigit()]

  return digits[0] if len(digits) else find_first_digit(line, reverse, pos+1) 

for i in range(len(data)):
  cals[i].update({ 'start': find_first_digit(data[i]) })
  cals[i].update({ 'end': find_first_digit(data[i][::-1], reverse=True) })

total = sum([int(c.get('start') + c.get('end')) for c in cals])
total
```
