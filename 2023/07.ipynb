```python
# read and parse

with open('07.txt', 'r') as f:
  hands = f.read().split('\n')

hands = [[hand.split(' ')[0], int(hand.split(' ')[1])] for hand in hands]
```

```python
# parts 1 & 2
# p1 correct answer: 246795406
# p2 correct answer: 249356515

def normal_group(groups, hand, l, m):
  if l == 1: groups[1] = groups.get(1, []) + [hand]
  elif l == 2 and m == 4: groups[2] = groups.get(2, []) + [hand]
  elif l == 2 and m == 3: groups[3] = groups.get(3, []) + [hand]
  elif l == 3 and m == 3: groups[4] = groups.get(4, []) + [hand]
  elif l == 3 and m == 2: groups[5] = groups.get(5, []) + [hand]
  elif l == 4: groups[6] = groups.get(6, []) + [hand]
  elif l == 5: groups[7] = groups.get(7, []) + [hand]

def joker_group(groups, hand, l, m, j): 
  if l == 1: groups[1] = groups.get(1, []) + [hand]
  elif l == 2 and m == 4: groups[1] = groups.get(1, []) + [hand]
  elif l == 2 and m == 3: groups[1] = groups.get(1, []) + [hand]
  elif l == 3 and m == 3: groups[2] = groups.get(2, []) + [hand]
  elif l == 3 and m == 2 and j == 1: groups[3] = groups.get(3, []) + [hand]
  elif l == 3 and m == 2 and j == 2: groups[2] = groups.get(2, []) + [hand]
  elif l == 4: groups[4] = groups.get(4, []) + [hand]
  elif l == 5: groups[6] = groups.get(6, []) + [hand] 

def sort(groups, joker=False):
  alphabet = 'AKQT987654432J' if joker else 'AKQJT98765432' # alphabet to use for sorting

  for i in range(1, len(groups)+1):
    groups[i].sort(key=lambda word: [alphabet.index(c) for c in word[0]])

  ranked_hands = [b := [], [b := b + a for a in list(groups.values())]][-1][-1][::-1]
  return ranked_hands

def calculate_total(hands, joker=False):
  groups = { i: [] for i in range(1, 8)} 

  for hand in hands:
    l, m, j = len(set(hand[0])), max(hand[0].count(c) for c in set(hand[0])), hand[0].count('J')
    joker_group(groups, hand, l, m, j) if joker and j > 0 else normal_group(groups, hand, l, m)

  ranked_hands = sort(groups, joker)
  return sum([(i+1)*ranked_hands[i][1] for i in range(len(hands))])

print('part 1:', calculate_total(hands))
print('part 2:', calculate_total(hands, joker=True))
```
