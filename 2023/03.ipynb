```python
# read

with open('03.txt', 'r') as f:
  data = f.read().split('\n')

# add a perimeter of periods to make my life easier
data = ['.' + d + '.' for d in data]
data = ['.'*len(data[1])] + data + ['.'*len(data[1])]
```

```python
# part 1
# correct answer: 540025

def has_adjacent_symbol(i, j):
  positions = [[i-1, j-1], [i-1, j], [i-1, j+1], [i, j-1], [i, j+1], [i+1, j-1], [i+1, j], [i+1, j+1]]

  for x, y in positions:
    if (not data[x][y].isdigit() and data[x][y] != '.'): return True

def add_parts(data):
  digits, digit, cont = [], '', False

  for i in range(len(data)):
    for j in range(len(data[i])):
      if data[i][j].isdigit():
        digit += data[i][j]
        if has_adjacent_symbol(i, j): cont = True
      else:
        if cont and len(digit): digits.append(int(digit))
        digit, cont = '', False
  
  return sum(digits)

print(add_parts(data))
```

```python
# part 2
# correct answer: 84584891

def find_gear(i, j):
  positions = [[i-1, j-1], [i-1, j], [i-1, j+1], [i, j-1], [i, j+1], [i+1, j-1], [i+1, j], [i+1, j+1]]

  for x, y in positions:
    if data[x][y] == '*': return str(x) + ':' + str(y)


def add_ratios(data):
  gears, gear, digit = {}, [], ''

  for i in range(len(data)):
    for j in range(len(data[i])):
      if data[i][j].isdigit():
        digit += data[i][j]
        gear = gear or find_gear(i, j)
      else:
        if gear and digit: gears[gear] = (gears.get(gear) or []) + [int(digit)]
        digit, gear = '', []

  return sum([v[0]*v[1] if len(v) == 2 else 0 for v in gears.values()])

print(add_ratios(data))
```
