```python
# read & parse

with open('04.txt', 'r') as f:
  data = f.read().split('\n')

cards = {}
for k in data:
  k, v = k.split(':')
  cards[int(k.strip('Card '))] = [x.split(' ') for x in v.strip().split('| ')]
```

```python
# part 1
# correct answer: 23235

matches, total = {}, 0
for k in cards:
  winning_nums, card_nums = cards.get(k)
  num_matches = sum([1 for n in card_nums if n and n in winning_nums])
  matches[k] = num_matches
  total += 2**(num_matches-1) if num_matches > 0 else 0

total
```

```python
# part 2
# correct answer: 5920640

counts = {k: 0 for k in cards}

def recursive_total(k):
  counts[k] += 1
  for i in range(k+1, k+matches.get(k)+1): recursive_total(i)

for k in cards: recursive_total(k)

print(sum(list(counts.values())))
```
