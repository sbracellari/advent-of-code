```python
# read and parse

with open('06.txt', 'r') as f:
  data = f.read().split('\n')

data = [[x for x in [d.split(' ') for d in data][i] if x.isdigit()] for i in range(2)]

data
```

```python
# part 1
# correct answer: 861300

total = 1
for t, d in zip(data[0], data[1]):
  total *= sum([1 if i*(int(t)-i) > int(d) else 0 for i in range(1, int(t))])

total
```

```python
# part 2
# correct answer: 28101347

new_time, new_distance = int(''.join(t for t in data[0])), int(''.join(t for t in data[1]))

total = sum([1 if i*(new_time-i) > new_distance else 0 for i in range(1, new_time)])
total
```
