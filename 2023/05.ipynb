```python
# part 1
# correct answer: 825516882

locs, seeds = [], ()

with open('05.txt','r') as f:
  for line in f:
    seeds = [int(x) for x in line.strip().split(':')[1].split()]
    break

  for s in seeds:
    found = False
    for line in f:
      line = line.strip()
      if line == '' or "map" in line or "seeds" in line:
        found = False
        continue
      if found: continue
      
      dest, src, r = (int(x) for x in line.split())
      interval = (src, src+r-1)
      if s >= interval[0] and s <= interval[1]:
        diff = src - dest
        s -= diff
        found = True
    locs.append(s)
    f.seek(0)

print(min(locs))
```
