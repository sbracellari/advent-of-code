```python
# read and parse

with open('08.txt', 'r') as f:
  data = f.read().split('\n\n')

moves = data[0].replace('L', '0').replace('R', '1')
network = { x: y for x, y in [[x, tuple(y.replace('(', '').replace(')', '').split(', '))] for x,y in [d.split(' = ') for d in data[1].split('\n')]]}
```

```python
# part 1
# correct answer: 11911

count, i = 0, 0
current = 'AAA'

while (current != 'ZZZ'):
  next = network.get(current)[int(moves[i])]
  i = (i+1)%len(moves)
  count += 1
  current = next

print(count)
```
