```python
# read & parse

with open('02.txt', 'r') as f:
  data = f.read().split('\n')

games = {}
for game in data:
  k, v = game.split(':')
  games[int(k.strip('Game '))] = [x.strip().split(', ') for x in v.split(';')]

games.get(3)
```

```python
# part 1
# correct answer: 2176

cubes = { 'red': 12, 'green': 13, 'blue': 14 }

def validate_game(games, k):
  for round in games.get(k):
    for cube in round:
      number, color = int(cube.split(' ')[0]), cube.split(' ')[1]
      
      if cubes.get(color) >= number: continue
      else: return 0
   
  return k

total = sum([validate_game(games, k) for k in games])
total
  
```

```python
# part 2
# correct answer: 63700

def min_cubes(game):
  cubes = { 'red': 0, 'green': 0, 'blue': 0 }

  for round in game:
    for cube in round:
      number, color = int(cube.split(' ')[0]), cube.split(' ')[1]
      cubes.update({ color: max(cubes.get(color), number)})

  return [b := 1, [b := b * a for a in list(cubes.values())]][-1][-1]

total = sum([min_cubes(games.get(k)) for k in games])
total
```
