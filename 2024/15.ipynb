```python
Point = tuple[int, int]
Grid = dict[Point, str]

with open('15.txt', 'r') as f:
  data: str = f.read()

map: Grid = {(i, j): v for i, row in enumerate(data.split('\n\n')[0].split('\n')) for j, v in enumerate(row)}
pos: Point = next((k for k, v in map.items() if v == '@'), (0, 0))
moves: str = data.split('\n\n')[1].replace('\n', '')
directions: dict[str, Point] = { '^': (-1, 0), '>': (0, 1), '<': (0, -1), 'v': (1, 0)}

def move(a: Point, b: Point) -> Point:
  return a[0] + b[0], a[1] + b[1]

for m in moves:
  next_pos: tuple[int, int] = move(pos, directions[m])
  if map[next_pos] == '#': continue
  elif map[next_pos] == '.': 
    map[pos] = '.'
    map[next_pos] = '@'
    pos = next_pos
  else:
    positions: list[Point] = []
    while (np := move(next_pos, directions[m])) in map:
      positions.append(np)
      if map[np] == '.': 
        map[pos] = '.'
        n: Point = move(pos,directions[m])
        map[n] = '@'
        pos = n
        for p in positions: map[p] = 'O'
        break
      elif map[np] == 'O': next_pos = np
      else: break
  
coords: int = sum([(100*k[0]+k[1]) for k, v in map.items() if v == 'O'])
coords
```
