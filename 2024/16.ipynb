```python
dim: int = len(data.split('\n'))
maze: Grid = {(i, j): v for i, row in enumerate(data.split('\n')) for j, v in enumerate(row)}
start: Point = (dim-2, 1) # bottom left corner (inside wall)
direction = '>'
directions: dict[str, Point] = { '^': (-1, 0), '>': (0, 1), 'v': (1, 0), '<': (0, -1)}

def turn(direction: str, rot: int) -> str:
  keys: list[str] = list(directions.keys())
  return keys[(keys.index(direction) + rot) % 4]

def search(maze: Grid, p: Point) -> bool:
  if maze[p] == 'E': return True
  elif maze[p] == '#': return False
  elif maze[p] == 'X': return False

  maze[p] = 'X'
  y, x = p

  
  if (search(maze, (y, x+1))
      or search(maze, (y-1, x))
      or search(maze, (y, x-1))
      or search(maze, (y+1, x))): return True
  
  return False

search(maze, start)
```
